Before Mapping strings to numeric type for classififcation

```python
df_dropped_default = df.drop(columns=['default'])
df1 = df_dropped_default.iloc[:, 0:7]
df1
```

Mapped the housing & loan to numeric values for further classification without using OneHotEncoding/Ordinal_encoding

```python
df1['marital']=df1['marital'].map({'married':2,'single':1,'divorced':0})
df1['housing']=df1['housing'].map({'yes':1,'no':0})
df1['loan']=df1['loan'].map({'yes':1,'no':0})
df1=df1.drop(columns=['education','job'],axis=1)
df1.sample(3)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df1.drop('loan',axis=1),df1['loan'],test_size=0.2,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
sc.transform(x_train)
sc.transform(x_test)
```

ML models without using Encoding methods:

1.Logistic_Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
print('Logistic_Regression predicted training score',lr.score(x_train,y_train)*100)
print('Logistic_Regression predicted test score',lr.score(x_test,y_test)*100)
```

2.XGBoost

```python
from xgboost import XGBClassifier
xg=XGBClassifier()
xg=xg.fit(x_train,y_train)
print('XGBoost predicted training score',xg.score(x_train,y_train)*100)
print('XGBoost predicted test score',xg.score(x_test,y_test)*100)
```

3.Random_Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=150,max_depth=5)
rf=rf.fit(x_train,y_train)
print('Random Forest predicted training score',rf.score(x_train,y_train)*100)
print('Random Forest predicted test score',rf.score(x_test,y_test)*100)
```

Using Encoding techniques

```python
df2 = df.drop(df.iloc[:, 8:15].columns, axis=1)
df2.sample(3)
```

Used Column Transformer for using different encoding techniques.
^OneHotEncoder
^OrdinalEncoder

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(drop='first', sparse_output=False),
         ['job','marital', 'poutcome', 'card']),
        ('ordinal', OrdinalEncoder(categories=[
            ['unknown','primary','secondary','tertiary'],
            ['no','yes'],
            ['no','yes'],
            ['no','yes']
        ]), ['education','housing', 'default', 'y'])
      ],
    remainder='passthrough'
)
```

Splitting the data as training and test data for further calculations. Mapping the string values to numeric

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df2.drop(columns='loan',axis=1),df2['loan'],test_size=0.2,random_state=42)
Y_train = Y_train.map({'yes': 1, 'no': 0})
Y_test = Y_test.map({'yes': 1, 'no': 0})
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

ML Model using Column_Transform

1.LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

l=LogisticRegression()
l=l.fit(X_train, Y_train)
print('Logistic_Regression predicted training score',l.score(X_train,Y_train)*100)
print('Logistic_Regression predicted test score',l.score(X_test,Y_test)*100)
```

2.RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
r=RandomForestClassifier(n_estimators=150,max_depth=10)
r=r.fit(X_train,Y_train)
print('Random Forest predicted training score',r.score(X_train,Y_train)*100)
print('Random Forest predicted test score',r.score(X_test,Y_test)*100)
```

3.XGBoost

```python
from xgboost import XGBClassifier
x=XGBClassifier()
x.fit(X_train,Y_train)
print('XGBoost predicated training score',x.score(X_train,Y_train)*100)
print('XGBoost predicated test score',x.score(X_test,Y_test)*100)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_excel('/content/bank_data.xlsx')
```

```python
y = df['loan'].replace({'no': 0, 'yes': 1})
```

```python
y = df['loan'].replace({'no': 0, 'yes': 1})
x = df.drop(['loan'], axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=43)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

ct = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown='ignore'), ['job']),
    ('ord', OrdinalEncoder(categories=[
        ['divorced', 'married', 'single'],
        ['secondary','primary','tertiary','unknown'],
        ['no','yes'],
        ['no','yes']
    ]), ['marital', 'education', 'default', 'housing']),
 ('num', MinMaxScaler(), ['age', 'balance', 'loan stats', 'Loan_period'])

], remainder='passthrough')
```

```python
df.sample()
```

```python
from sklearn.feature_selection import SelectKBest, chi2
ct4=SelectKBest(score_func=chi2,k=12)
```

1.RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

2.LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
lgr=LogisticRegression()
```

3.XGBClassifier

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
```

Creating pipeline for the models

```python
pipe=Pipeline([
    ('ct',ct),
    ('ct4',ct4),
    ('lgr',lgr)
])
```

```python
pipe=pipe.fit(x_train,y_train)
```
